# Cellular Automata

[Cellular automata](https://en.wikipedia.org/wiki/Cellular_automaton) are simple discrete dynamical systems. They consist of a grid of cells that can each be on or off. At each timestep, a cell is turned on or off depending on which cells in its neighborhood (including itself) are on or off. Despite having extremely simple rules determining their behavior, they demonstrate a number of important concepts in dynamical systems theory, like self-organization, emergence, chaos, and attractors. And, they can be used to model important real-word phenomena like [the evolution of a society of organisms](https://en.wikipedia.org/wiki/Conway%27s_Game_of_Life).

So let's play with cellular automata a little bit. For simplicity, let's start with elementary cellular automata, for which the grid of cells is one-dimensional. There are many interesting rulesets for elementary cellular automata, but my favorite so far is Wolfram's Rule 30, which seems to generate the pattern on the <i>Conus textile</i>'s shell.

```python
import numpy as np
import random
import seaborn as sns
```

```python
ruleThirty=  """
             inputs: 111	110	101	100	011	010	001	000
             outputs: 0	 0	 0	 1	 1	 1	 1	 0
             """
inputs = ruleThirty.split('\n')[1]
inputs = inputs.split()[1:]
print(inputs)

outputs = ruleThirty.split('\n')[2]
outputs = outputs.split()[1:]
print(outputs)

ruleThirty = {tuple( int(z) for z in x ):int(y) for x, y in zip(inputs,outputs)}

#see for yourself that ruleThirty is structured appropriately
for inputs, output in ruleThirty.items():
    print(inputs,output)

#def ruleThirtyFunc(neighborhood): return ruleThirty[neighborhood]
#print(ruleThirtyFunc((1,0,0)))
```

We're almost ready to try evolving the grid. First, however, we have to decide how to handle edge cells. Consider the rightmost cell: it doesn't have a rightside neighbor, so how does it figure out how to evolve? At least a few ways to deal with this: (1) infinite grid, (2) edges are constant, (3) wrap the grid around. We'll opt for #3.

Let's try to evolve the grid one time and see what happens.

```python
"""
def evolve(grid, rules, edge = 'wrap'):
    middle    = np.array(grid)
    bottom    = np.roll(grid, -1)
    top       = np.roll(grid,  1)
    newGrid   = np.array(grid)
    # we should really vectorize this. one option would be to convert newGrid to a len(newGrid) x 3 matrix, and then
    # simultaneously apply the rules dict/function to all of them?
    for cellInd, cellVal in enumerate(grid):
        neighborhood     = tuple(grid.take([cellInd - 1, cellInd, cellInd + 1], mode = edge)) # note wrap mode
        newCellVal       = rules[neighborhood]
        newGrid[cellInd] = newCellVal
    return newGrid
"""

def evolve(grid, rules, edge = 'wrap'):
    grid    = np.array(grid)
    newGrid = np.array(grid)
    # we should really vectorize this. one option would be to convert newGrid to a len(newGrid) x 3 matrix, and then
    # simultaneously apply the rules dict/function to all of them?
    for cellInd, cellVal in enumerate(grid):
        neighborhood     = tuple(grid.take([cellInd - 1, cellInd, cellInd + 1], mode = edge)) # note wrap mode
        newCellVal       = rules[neighborhood]
        newGrid[cellInd] = newCellVal
    return newGrid

gridSize = 20
startGrid = np.array( [random.randint(0,1) for x in range(gridSize)] )

print("At time 1:", startGrid)
newGrid = evolve(startGrid, ruleThirty)
print("At time 2:", newGrid)
```

It appears to have worked! Let's try running it for longer periods and seeing what happens.

```python
steps = 20

def evolveMany(grid, steps, rules, evolvingFxn):
    """
    Evolves a 1d cellular automata grid steps times, using rules and evolvingFxn.
    
    Returns a len(grid) x steps matrix with the history of the CA.
    """
    
    gridMat = np.zeros(shape=(steps,len(grid)), dtype=np.int8)
    gridMat[0] = grid
    for x in range(steps - 1):
        #if not x % 100:
        #    print(grid)
        #print(gridMat[x])
        gridMat[x+1] = evolvingFxn(gridMat[x], rules)
    return gridMat

evolveMany(startGrid, steps, ruleThirty, evolve)
```

When you start with random inputs, it's kind of hard to see the structure in its evolution. Let's start with some simple inputs, and see where it leads us. Let's try just a single 1 in the middle of all 0's.

```python
gridSize = 20
startGrid = np.array([0 for x in range(gridSize)])
startGrid[gridSize/2] = 1

steps = 20
evolveMany(startGrid, steps, ruleThirty, evolve)
```

Okay, that's looking like the classic rule 30 pattern. But it's a little harder to see. What might be a better way to visualize? Let's go for the classic black-and-white square visualization of a CA. And what the hell -- how about we hugely expand the gridsize and number of steps, to 1000 each. This will let us see some really interesting stuff.

```python
%pylab inline
pylab.rcParams['figure.figsize'] = (30, 24)

gridSize  = 1000
steps     = 1000
startGrid = np.array([0 for x in range(gridSize)])
startGrid[gridSize / 2] = 1
cellHistory = evolveMany(startGrid, steps, ruleThirty, evolve)

sns.heatmap(cellHistory, cmap = 'Greys', cbar=False, square = True, xticklabels = False, yticklabels = False);
```

That's a spicy meatball. So now we can see a few of the concepts we alluded to at the beginning First, even though the rules are extremely simple, the resulting evolution of the grid shows remarkable complexity. This is what we referred to earlier as **emergence** -- complex, large-scale behavior from simple dynamics/rules. While the left-side of the resulting pyramid shows a great deal of regularity and apparent structure, the rest of the pyramid displays *seeming* random behavior. I say 'seeming', because **it is NOT, in fact, random: the rules of evolution are deterministic!** For example, for rule 30, if a cell and its neighbors are all on, then it *always* turns off on the next time step. Finally, it can be shown that the behavior of the rule 30 CA qualifies as **chaotic** (if you compare the evolution of one particular starting grid to the evolution of an arbitrarily similar starting grid, you will see that they rapidly diverge).

You may be wondering how Rule 30 would generate the shell we saw above. From [Wikipedia](https://en.wikipedia.org/wiki/Cellular_automaton#Biology):

> The pigment cells reside in a narrow band along the shell's lip. Each cell secretes pigments according to the activating and inhibiting activity of its neighbor pigment cells, obeying a natural version of a mathematical rule. The cell band leaves the colored pattern on the shell as it grows slowly.

Now, let's try a different ruleset. Let's try Rule 110.

```python
ruleOneTen=  """
             inputs: 	111	110	101	100	011	010	001	000
             outpus:      0	 1	 1	 0	 1	 1	 1	 0
             """
inputs = ruleOneTen.split('\n')[1]
inputs = inputs.split()[1:]
outputs = ruleOneTen.split('\n')[2]
outputs = outputs.split()[1:]
ruleOneTen = {tuple( int(z) for z in x ):int(y) for x, y in zip(inputs,outputs)}

gridSize = 1000
steps    = 1000
startGrid = np.array([0 for x in range(gridSize)])
startGrid[-1] = 1
cellHistory = evolveMany(startGrid, steps, ruleOneTen, evolve)

%pylab inline
pylab.rcParams['figure.figsize'] = (30, 24)
sns.heatmap(cellHistory, 
            cmap = 'Greys', 
            cbar=False, 
            square = True, xticklabels = False, yticklabels = False);
```

Now we see mostly regular, rather elaborate structure, but for that descending line on the right.

It also turns out that Rule 110 is **Turing complete**, meaning it can simulate an arbitrary Turing Machine. In other words, any calculation or computer program can be simulated using a rule 110 automaton.
