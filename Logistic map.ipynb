# Logistic map

The [logistic map](https://en.wikipedia.org/wiki/Logistic_map) is a well-known simple, discrete dynamical system of the form:

![](https://wikimedia.org/api/rest_v1/media/math/render/svg/66f02b84769935f1d57fccf187e40e76e9db4570)

The map was originally made well-known as a model of population dynamics -- how a population grows or shrinks over time. The value `x_n` is a number between zero and one that represents the ratio of existing population to the maximum possible population. When the parameter `r` is restricted to [0,4], the map can capture important population growth and decline dynamics.

Let's explore the behavior of the logistic map a little bit.

```python
import numpy as np
import matplotlib.pyplot as plt
import time, os
import matplotlib.cm as cm

folder = '/Users/russellrichie/dynamical_systems_demos/'
os.chdir(folder)
```

```python
xa=0
xb=4
breakTime = 1

R = np.linspace(xa,xb,1000)
print("Your parameter space is:\n", R)
```

```python
Y = R*0+0.5
print("Your initial conditions are: \n", Y)
```

Evolve the dynamical system for `initialMapCount` maps/iterations, printing off the current `Y` every now and then. By mapping ~1000 times, we get past the initial conditions

```python
initialMapCount = 1000
logistic_map = lambda y, r: r*y*(1-y)
for x in range(initialMapCount):
    #if not x % 200:
    #    time.sleep(breakTime)
    #    print("After {} maps, Y is: \n".format(x), Y, "\n")
    #Y = Y**2 - R
    #Y = R*Y*(1-Y)
    Y = logistic_map(Y,R)
```

## Building a bifurcation diagram

Now plot the long-term stabilities of the map. By continuing to evolve the system by `plottingMapCount`, we ensure that, in the ranges of `C` with multiple stabilities, we hit all those stabilities and plot them. For `C` < 3, there is only one stability, so this continued iteration doesn't do anything. But for `C` > 3, there are multiple stabilities, so continued iteration flips back and forth between these.

```python
%pylab inline
pylab.rcParams['figure.figsize'] = (30, 24)

plottingMapCount = 1000
markerSize       = 0.05
for y in range(plottingMapCount):
    #if not y % 100:
    #    Y = Y**2 - R
    #Y = Y**2 - R
    #Y = R*Y*(1-Y)
    Y = logistic_map(Y,R)
    plt.scatter(R, Y, s=markerSize)

plt.xlabel('r')
plt.ylabel('y')
plt.show()
```

So we can now see a couple things. First, for 0 < r < 1, the map converges on 0. If we were modeling a population, this would mean that the population would die out! For 1 < r, 3, the map converges on a positive value which increases as a function of r. Things get interesting for 3 < r < 4: we see bifurcation -- splitting of the stabilities, and eventually, around r = 3.56, we see descent into chaos. A full descriptoin of the behaviors can be found [here](https://en.wikipedia.org/wiki/Logistic_map#Behavior_dependent_on_.7F.27.22.60UNIQ--postMath-00000007-QINU.60.22.27.7F).

## Building a cobweb plot

Now build cobweb plots for particular values of parameter r. Again, assuming `y_0` = 0.5. Basically, recreate this:

![](https://upload.wikimedia.org/wikipedia/commons/e/ed/LogisticCobwebChaos.gif)

Let's break that down a little bit. Each image in that gif is something like this:

![](https://upload.wikimedia.org/wikipedia/commons/4/41/CobwebConstruction.gif)

What *this* gif is showing is how the map evolves from a particular starting point (x~.8), for a particular parameter (r~2.7). [Wikipedia explains it pretty well](https://en.wikipedia.org/wiki/Cobweb_plot):
1. Find the point on the function curve with an x-coordinate of x_0.
    This has the coordinates (x_0, f(x_0)).
2. Plot horizontally across from this point to the diagonal line. 
    This has the coordinates (f(x_0), f(x_0)).
3. Plot vertically from the point on the diagonal to the function curve. 
    This has the coordinates (f(x_0), f(f(x_0))).
4. Repeat from step 2 as required.

**In other words, successive points on the curve plotting the map represent successive iterations of the map.**

So what we're seeing in the above plot is the map converging on ~.63. For this parameter, that is the attractor.

So the top gif, then, shows how the evolution and the attractors change as we vary the parameter. We can see that crossing certain parameter values drastically changes the evolution and overall behavior of the map.

So now let's work on re-creating the cobweb diagram.

```python
# we'll use x_min and x_max for two things:
# 1) to establish the range of the axes on the plot
# 2) to make xs. we'll plot every f(xs) so we can visualize the mapping function

x_min = 0
x_max = 1
xs = np.linspace(x_min, x_max, 1000)
print(xs)
```

```python
paramCount = 100
R  = np.linspace(xa, xb, paramCount)
print("Your parameter space is:\n", R)
```

```python
y = 0.5
print("The map will start at x =", y)
#r = R[0]
```

```python
folder = '/Users/russellrichie/dynamical_systems_demos/cobwebs'
os.chdir(folder)

markerSize = 0.1
for index, r in enumerate(R):
    rLabel = 'r = {0:.2f}'.format(r)
    if not index % 10:
        print("Currently cobwebbing {}, {} of {} parameters".format(rLabel, 
                                                                index + 1,
                                                                len(R)
                                                                )
              )
    plt.xlim(x_min,x_max)
    plt.ylim(x_min,x_max)
    
    # plot y = x
    plt.plot(range(x_max + 1), linewidth = markerSize*10, c = 'blue')
    
    # plot the mapping function
    plt.scatter(xs, logistic_map(xs,r), s = markerSize)
    #plt.scatter(xs, r*xs*(1-xs), s = markerSize)
    #plt.scatter(xs,  xs**2 - r, s = markerSize)

    # now, plot cobweb diagrams for first initialMapCount iterations
    #y = 1
    y = 0.5
    #point1 = (y, y**2 - r)
    #point1 = (y, r*y*(1-y))
    point1 = (y, logistic_map(y,r))
    points = [point1]
    initialMapCount = 100
    for _ in range(initialMapCount):
        #y = y**2 - r
        #y = r*y*(1-y)
        y = logistic_map(y,r)
        point2 = (y, y) 
        #point3 = (y, y**2 - r)
        point3 = (y, logistic_map(y,r))
        points.extend([point2,point3])
    points = np.array(points)
    colors = cm.hot(np.linspace(0, 1, len(points)))
    plt.plot(points[:,0], 
             points[:,1], 
             linewidth = markerSize*6,
             color = 'red', 
             linestyle = 'solid',
             label = rLabel
             )
    plt.legend(loc='upper left')
    plt.ylabel('y')
    plt.xlabel('x')
    cobwebFilename = 'logistic cobweb for ' + rLabel + '.png'
    plt.savefig(filename = cobwebFilename)
    plt.clf()
print("Done making cobwebs.")
os.chdir('/Users/russellrichie/dynamical_systems_demos/')
```
